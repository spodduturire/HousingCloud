```python
import numpy as np
import pandas as pd
```

```python
inventory = pd.read_csv('inventory_data.csv')
inventory
```

```python
occupancy = pd.read_csv('occupancy_data.csv')
occupancy
```

```python
person_bedid = pd.merge(occupancy, inventory, on=['roomName', 'bedName'], how='inner')
person_bedid = person_bedid[['personId', 'bedId']]
person_bedid
```

```python
majors = pd.read_csv('majors_data.csv')
majors.drop_duplicates(subset=['name'], keep='first', inplace=True)
majors['name'] = majors['name'].str.strip()
major_name_id_dic = {}

for idx, row in majors.iterrows():
  major_name_id_dic[row['name']] = row['id']
```

```python
persons = pd.read_csv('persons_data.csv')
persons.drop_duplicates(subset=['email'], keep='first', inplace=True)
persons
```

```python
persons['name'] = persons['firstName'].str.cat(persons['lastName'], sep=' ')
persons = persons[['personId', 'name', 'email', 'dob', 'address','majors']]
persons
```

```python
persons[['address1', 'city', 'state']] = persons['address'].str.split(',', expand=True)
persons['address2'] = ''
persons['zip'] = ''
persons = persons[['personId', 'name', 'email', 'dob', 'address1', 'address2', 'city', 'state', 'zip', 'majors']]
persons
```

```python
def get_major_id(major_names):
  if major_names == '':
    return ''
  majorIds = ''
  major_lst = major_names.split(',')
  for major in major_lst:
    major = major.strip()
    majorIds += major_name_id_dic[major] + ', '
  return majorIds[:-2]
```

```python
persons['majors'] = persons['majors'].fillna('')
```

```python
persons['majorIds'] = persons['majors'].apply(get_major_id)
persons = persons[['personId', 'name', 'email', 'dob', 'address1', 'address2', 'city', 'state', 'zip', 'majorIds']]
persons
```

```python
final_csv = pd.merge(persons, person_bediid, on='personId', how='left')
final_csv['bedId'] = final_csv['bedId'].fillna('')
final_csv
```

```python
final_csv.to_csv('final_data_python.csv', index=False)
```
